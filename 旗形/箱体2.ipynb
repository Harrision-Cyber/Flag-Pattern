```python
def find_pips(data: np.array, n_pips: int, dist_measure: int):
    """
    感知重要点(Perceptually Important Points, PIP)算法
    
    参数:
    data: 价格数据数组
    n_pips: 要识别的重要点数量
    dist_measure: 距离度量方式
        1 = 欧几里得距离(Euclidean Distance)
        2 = 垂直距离(Perpendicular Distance)
        3 = 垂直距离(Vertical Distance)
    
    返回:
    pips_x: 重要点的索引
    pips_y: 重要点的价格值
    """
    pips_x = [0, len(data) - 1]
    pips_y = [data[0], data[-1]]

    for curr_point in range(2, n_pips):
        md = 0.0
        md_i = -1
        insert_index = -1

        for k in range(0, curr_point - 1):
            left_adj = k
            right_adj = k + 1

            time_diff = pips_x[right_adj] - pips_x[left_adj]
            price_diff = pips_y[right_adj] - pips_y[left_adj]
            slope = price_diff / time_diff
            intercept = pips_y[left_adj] - pips_x[left_adj] * slope

            for i in range(pips_x[left_adj] + 1, pips_x[right_adj]):
                d = 0.0
                
                if dist_measure == 1:
                    d = ((pips_x[left_adj] - i) ** 2 + (pips_y[left_adj] - data[i]) ** 2) ** 0.5
                    d += ((pips_x[right_adj] - i) ** 2 + (pips_y[right_adj] - data[i]) ** 2) ** 0.5
                elif dist_measure == 2:
                    d = abs((slope * i + intercept) - data[i]) / (slope ** 2 + 1) ** 0.5
                else:
                    d = abs((slope * i + intercept) - data[i])

                if d > md:
                    md = d
                    md_i = i
                    insert_index = right_adj
        pips_x.insert(insert_index, md_i)
        pips_y.insert(insert_index, data[md_i])

    return pips_x, pips_y
```

```python
# 计算箱体边界
def calculate_box_boundaries(pips_y, window_size=100, method='percentile'):
    """
    计算箱体的上下边界
    
    参数:
    pips_y: array - PIP点的价格值
    window_size: int - 滑动窗口大小,用于计算局部箱体
    method: str - 计算边界的方法,可选值:'percentile','std','minmax','bollinger'
    
    返回:
    upper_bound: float - 箱体上边界
    lower_bound: float - 箱体下边界
    """
    
    
    if method == 'percentile':
        # 分位数法
        upper_bound = np.percentile(pips_y, 80)  
        lower_bound = np.percentile(pips_y, 20)
    elif method == 'std':
        # 标准差法
        mean = np.mean(pips_y)
        std = np.std(pips_y)
        upper_bound = mean + 1 * std  # 均值上下2个标准差
        lower_bound = mean - 1 * std
    elif method == 'minmax':
        # 最大最小值法
        upper_bound = np.max(pips_y)
        lower_bound = np.min(pips_y)
    elif method == 'bollinger':
        # 布林带法
        rolling_mean = np.mean(pips_y)
        rolling_std = np.std(pips_y)
        upper_bound = rolling_mean + 2 * rolling_std
        lower_bound = rolling_mean - 2 * rolling_std
    
    return upper_bound, lower_bound

# 判断价格是否在箱体内
def is_price_in_box(price, upper_bound, lower_bound, threshold=0.1):
    """
    判断价格是否在箱体范围内
    
    参数:
    price: float - 当前价格
    upper_bound: float - 箱体上边界
    lower_bound: float - 箱体下边界 
    threshold: float - 允许突破箱体的阈值比例
    
    返回:
    bool - 价格是否在箱体内
    """
    box_height = upper_bound - lower_bound
    upper_bound_threshold_value = upper_bound * threshold
    lower_bound_threshold_value = lower_bound * threshold
    

    return (price >= lower_bound - lower_bound_threshold_value) and (price <= upper_bound + upper_bound_threshold_value)

```

```python
# 检查所有股票是否在箱体内
def check_stocks_in_box(df, window_size=100, pip_count=100, dist_measure=3, box_threshold=0.1):
    """
    检查所有股票是否在箱体内
    
    参数:
    df: DataFrame - 归一化后的股票数据
    window_size: int - 滑动窗口大小
    pip_count: int - PIP点数量
    dist_measure: int - PIP点距离度量
    box_threshold: float - 箱体突破阈值
    
    返回:
    dict - 每只股票是否在箱体内的结果
    """
    results = {}
    # 设定起始日期
    # start_date = '2024-09-24'

    # 遍历每只股票
    for stock_code in df.columns:
        # 获取股票数据并去除NaN值
        stock_data = df[stock_code].dropna().values
        
        # 如果数据点太少则跳过
        if len(stock_data) < window_size:
            continue
       
        
        # 获取2024.09.24之前的数据
        # historical_data = full_data[stock_data.index < start_date]

        # 计算PIP点
        _, pips_y = find_pips(stock_data, pip_count, dist_measure)


        # 计算箱体边界
        upper_bound, lower_bound = calculate_box_boundaries(pips_y, window_size)
        
        # 计算价格在箱体内的比例
        prices_in_box = sum(1 for p in stock_data if is_price_in_box(p, upper_bound, lower_bound, box_threshold))
        box_ratio = prices_in_box / len(stock_data)
        
        # 获取最新价格
        current_price = stock_data[0]
        
        # 判断是否在箱体内(比例超过70%)
        in_box = box_ratio > 0.7
        
        results[stock_code] = {
            'in_box': in_box,
            'current_price': current_price,
            'upper_bound': upper_bound,
            'lower_bound': lower_bound,
            'box_ratio': box_ratio
        }
        
    return results

# 执行分析
box_results = check_stocks_in_box(normalized_df)

# 统计在箱体内的股票(比例>70%的股票)
stocks_in_box = {k: v for k, v in box_results.items() if v['in_box']}
print(f"共有 {len(stocks_in_box)} 只股票处于箱体内")


# 显示在箱体内的股票详情
print("\n箱体内股票列表:")
for stock, info in stocks_in_box.items():
    current_price = normalized_df[stock].iloc[0]
    print(f"{stock}: 当前价格={current_price:.4f}, 上边界={info['upper_bound']:.4f}, 下边界={info['lower_bound']:.4f}, 箱体内比例={info['box_ratio']:.2%}")


```

```python
def check_box_volatility(stock_data, upper_bound, lower_bound):
    """
    计算箱体的波动率
    """
    # 计算箱体高度
    box_height = upper_bound - lower_bound
    
    # 计算价格的标准差
    price_std = np.std(stock_data)
    
    # 计算波动率(箱体高度与价格标准差的比值)
    volatility_ratio = box_height / price_std
    
    return volatility_ratio

def check_breakout_frequency(stock_data, upper_bound, lower_bound, threshold=0.1):
    """
    计算价格突破的频率和幅度
    """
    # 计算突破次数
    breakouts = 0
    box_height = upper_bound - lower_bound
    max_breakout = 0
    
    # or price < lower_bound * (1 - threshold)

    for price in stock_data:
        if price > upper_bound * (1 + threshold) :
            breakouts += 1
            # 计算突破幅度
            upper_breakout = max(0, price - upper_bound)
            lower_breakout = max(0, lower_bound - price)
            max_breakout = max(max_breakout, upper_breakout)
    
    # 计算突破比例
    breakout_ratio = breakouts / len(stock_data)
    
    # 计算最大突破幅度占箱体高度的比例
    max_breakout_ratio = max_breakout / box_height if box_height > 0 else 0
    
    return breakout_ratio, max_breakout_ratio

def calculate_slope(stock_data):
    """
    计算价格走势的斜率
    """
    # 生成时间序列索引
    x = np.arange(len(stock_data))
    # 将stock_data转换为numpy数组并确保为float类型
    y = np.array(stock_data.tolist(), dtype=float)
    # 反转数组使最新值在最后
    y = y[::-1]
    # 计算线性回归
    slope, _ = np.polyfit(x, y, 1)
    return slope

# 计算每只股票的波动率和突破频率
print("\n波动率、突破频率和斜率分析:")
for stock_code, info in stocks_in_box_filtered.items():
    # 获取该股票数据
    stock_data = normalized_df[stock_code].dropna()
    if len(stock_data) == 0:
        continue
        
    # 计算波动率
    volatility = check_box_volatility(stock_data, info['upper_bound'], info['lower_bound'])
    
    # 计算突破频率和最大突破幅度
    breakout_ratio, max_breakout_ratio = check_breakout_frequency(
        stock_data, 
        info['upper_bound'], 
        info['lower_bound']
    )
    
    # 计算斜率
    slope = calculate_slope(stock_data)
    
    print(f"\n{stock_code}:")
    print(f"波动率: {volatility:.4f}")
    print(f"突破次数占比: {breakout_ratio:.2%}")
    print(f"最大突破幅度占箱体高度的比例: {max_breakout_ratio:.2%}")
    print(f"价格走势斜率: {slope:.6f}")
```

```python
# 显示在箱体内的股票详情和分析
print("\n箱体内股票列表及分析:")
for stock, info in stocks_in_box_filtered.items():
    current_price = normalized_df[stock].iloc[0]
    # 计算波动率
    volatility = check_box_volatility(normalized_df[stock].dropna(), info['upper_bound'], info['lower_bound'])
    # 计算突破频率和最大突破幅度
    breakout_ratio, max_breakout_ratio = check_breakout_frequency(
        normalized_df[stock].dropna(),
        info['upper_bound'],
        info['lower_bound']
    )
    print(f"\n{stock}:")
    print(f"当前价格={current_price:.4f}, 上边界={info['upper_bound']:.4f}, 下边界={info['lower_bound']:.4f}, 箱体内比例={info['box_ratio']:.2%}, 波动率={volatility:.4f}, 突破频率={breakout_ratio:.2%}, 最大突破幅度占箱体高度比例={max_breakout_ratio:.2%}, 价格走势斜率={slope:.6f}")

# 创建一个字典来存储所有分析结果
analysis_results = []

for stock_code, info in stocks_in_box_filtered.items():
    stock_data = normalized_df[stock_code].dropna()
    if len(stock_data) == 0:
        continue
        
    # 计算各项指标
    volatility = check_box_volatility(stock_data, info['upper_bound'], info['lower_bound'])
    breakout_ratio, max_breakout_ratio = check_breakout_frequency(
        stock_data,
        info['upper_bound'], 
        info['lower_bound']
    )
    current_price = normalized_df[stock_code].iloc[0]

    slope = calculate_slope(stock_data)
    
    # 将结果添加到列表中
    analysis_results.append({
        '股票代码': stock_code,
        '当前价格': current_price,
        '上边界': info['upper_bound'],
        '下边界': info['lower_bound'],
        '箱体内比例': info['box_ratio'],
        '波动率': volatility,
        '突破次数占比': breakout_ratio,
        '最大突破幅度占箱体高度的比例': max_breakout_ratio,
        '价格走势斜率': slope
    })

# 转换为DataFrame并保存到Excel
import pandas as pd
df_results = pd.DataFrame(analysis_results)
df_results.to_excel('【2022年以来-斜率】箱体分析结果.xlsx', index=False)
print("\n分析结果已保存至'箱体分析结果1.xlsx'")

```
